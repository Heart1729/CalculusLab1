```python
# импорт необходимых библиотек для лабораторной работы
import numpy as np
import matplotlib.pyplot as plt
from math import sqrt
```

```python
def logistic_display(r_vals):
  """Строит график логистического отображения с заданными r_vals"""
  # Параметры
  x = np.linspace(0, 1, 400)

  plt.figure(figsize=(8, 6))

  for r in r_vals:
      y = r * x * (1 - x)
      plt.plot(x, y, label=f'$r = {r}$')

  # y = x — для удобства
  plt.plot(x, x, 'k--', linewidth=1, label='$y = x$')

  # Оформление
  plt.xlim(0, 1)
  plt.ylim(0, 1)
  plt.xlabel('$x_n$')
  plt.ylabel('$x_{n+1}$')
  plt.title('Логистическое отображение $x_{n+1} = r x_n (1 - x_n)$ при $r \\in (0, 1]$')
  plt.legend()
  plt.grid(True, linestyle='--', alpha=0.6)
  plt.axhline(0, color='black', linewidth=0.5)
  plt.axvline(0, color='black', linewidth=0.5)
  plt.show()

r_vals = [0.1, 0.3, 0.5, 0.8, 1.0]
logistic_display(r_vals)
```

```python
def spot_display(r_vals):
  """Строит график точечного отображения для N = 4 с заданными r_vals"""
  # Параметры
  x = np.linspace(0, 1, 400)

  plt.figure(figsize=(8, 6))

  for r in r_vals:
      y = r * x * (2 - x) * (1 - x)
      plt.plot(x, y, label=f'$r = {round(r, 3)}$')

  # y = x — для удобства
  plt.plot(x, x, 'k--', linewidth=1, label='$y = x$')

  # Оформление
  plt.xlim(0, 1)
  plt.ylim(0, 1)
  plt.xlabel('$x_n$')
  plt.ylabel('$x_{n+1}$')
  plt.title('Точечное отображение $x_{n+1} = r x_n (2 - x_n) (1 - x_n)$ при $r \\in [0, 3\sqrt{3} / 2]$')
  plt.legend()
  plt.grid(True, linestyle='--', alpha=0.6)
  plt.axhline(0, color='black', linewidth=0.5)
  plt.axvline(0, color='black', linewidth=0.5)
  plt.show()

r_vals = [0.0, 0.5, 1.0, 1.5, 2.0, 3 * sqrt(3) / 2]
spot_display(r_vals)
```
